### Campionamento del modulo quadro

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

from scipy.optimize import curve_fit
```

##### Temperatura

```python
figura, plot = plt.subplots(1, 1, figsize=(15, 6))
temp_sol = np.loadtxt("Osservabili/Solido/Temp.dat", unpack=True)

plt.errorbar(temp_sol[0, 0 : 50], temp_sol[2, 0 : 50], yerr = temp_sol[3, 0 : 50], label = 'Monte-Carlo', fmt = 'o-', color = 'blue')
plt.errorbar(temp_sol[0, 50 : 100], np.array([0.8 for i in range(0, 50)]), color = 'orange', label = 'Valor vero')
plt.errorbar(temp_sol[0, 50 : 100], temp_sol[2, 50 : 100], yerr = temp_sol[3, 50 : 100], label = 'Molecular-Dynamics', fmt = 'o-', color = 'red')

plt.grid(True)
plt.xlabel('Numero blocco')
plt.ylabel('Temperatura')
plt.legend(loc=(1.05, 0.8), frameon=False)  

plt.title('Temperatura: contronto NVE vs NVT', fontsize = '14')
```
